```python
#리스트에 20대확진자 수 80대 확진자 수 만듬
with open('TimeAge.csv',encoding ='cp949') as f:
    data = csv.reader(f)
    next(data)
    tw = []
    eg = []
    may = []
    for row in data :
        if row[0].split('-')[1] == '05':
            if row[2] == '20s' :
                tw.append(row[3])
                may.append(row[0].split('-')[-1])
            elif row[2] == '80s' :
                eg.append(row[3])
            
print('20대',tw)
print('80대',eg)
print('날짜',may)
```

```python
# 시각화 해보기 
plt.figure(figsize=(15,10))
plt.title('5월 20대와80대 확진자수')
plt.plot(may,tw,'blue', label='male')
plt.plot(may,eg,'red', label='female')
plt.legend()
plt.show()
```

## 20대와 80 사망자 수 

```python
with open('TimeAge.csv',encoding ='cp949') as f:
    data = csv.reader(f)
    next(data)
    tw2 = []
    eg2 = []
    may2 = []
    for row in data :
        if row[0].split('-')[1] == '05':
            if row[2] == '20s' :
                tw2.append(row[4])
                may2.append(row[0].split('-')[-1])
            elif row[2] == '80s' :
                eg2.append(row[4])
            
print('20대',tw2)
print('80대',eg2)
print('날짜',may2)
```

```python
# 시각화 해보기 
plt.figure(figsize=(15,10))
plt.title('5월 20대와80대 확진자수')
plt.plot(may2,tw2,'blue', label='male')
plt.plot(may2,eg2,'red', label='female')
plt.legend()
plt.show()
```

```python
#리스트에 남자확진자수와 여자확진자수로 나눔!
with open('TimeGender.csv',encoding ='cp949') as f:
    data = csv.reader(f)
    next(data)
    male = []
    female = []
    for row in data :
        if row[2] == 'male' and row[3] :
            male.append(row[3])
        elif row[2] == 'female' and row[3] :
            female.append(row[3])

print('남자',male)
print('여자',female)
```

```python
# 월별로 남자/여자 확진자 수 만들어 보기
with open('TimeGender.csv',encoding ='cp949') as f:
    data = csv.reader(f)
    next(data)
    male = []
    female = []
    mt = []
    for row in data :
        if row[2] == 'male' and row[3] :
            male.append(row[3])
        elif row[2] == 'female' and row[3] :
            female.append(row[3])
        elif row[0] == 'yy-mm-xx' :
            mt.append(row[0])
            
print('남자',male)
print('여자',female)
print('날짜', mt)
```
